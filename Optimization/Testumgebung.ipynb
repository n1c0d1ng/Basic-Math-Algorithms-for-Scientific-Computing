```python
import logistic_ODE as Model
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('testdaten.csv', delimiter=',', skiprows=1)
t_data = data[:,0]
x_data = data[:,1]
initial_params = [2.0, 2.0, 15.0] 
objective_function_analytic = Model.LogisticODE(t_data, x_data, analytic=True)
objective_function_numerical = Model.LogisticODE(t_data, x_data, analytic=False)
analytical_solution = objective_function_analytic.optimize(initial_params)
numerical_solution = objective_function_numerical.optimize(initial_params)
print("Analytical solution:", analytical_solution)
print("Numerical solution:", numerical_solution)
print("Loss (numerical):", objective_function_numerical.L(*numerical_solution))
print("Starting Loss (initial):", objective_function_numerical.L(*initial_params))
```

```python
params = [2.0, 2.0, 15.0]
analytical_solution = objective_function_analytic.x_analytic(*params)
plt.plot(t_data, x_data, 'o', label='Data', markersize=3, color="#dbf3f3")
plt.plot(t_data, analytical_solution, '-', label='Initial guess', color="#a9e844")
plt.plot(t_data, objective_function_numerical.x_simulate(*numerical_solution), '-', label='Fitted Numerical Solution', color="#ef543c")
plt.legend(facecolor='#24272f',labelcolor="#e7f3f3",frameon=False)
plt.gcf().set_facecolor('#24272f')  # Figure-Hintergrund
plt.gca().set_facecolor('#24272f')  # Achsen-Hintergrund
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color("#e7f3f3") # x-Achse Farbe
plt.gca().spines['left'].set_color("#e7f3f3") # y-Achse Farbe
plt.tick_params(axis='both', colors="#e7f3f3") # Ticks Farbe
plt.show()
```

```python
# Erstellung der Plots
true_params = [1.0, 0.5, 10.0]
plt.plot(t_data, x_data, 'o', label='Data', markersize=3, color="#dbf3f3")
plt.plot(t_data, objective_function_analytic.x_analytic(*true_params), '-', label='Analytic Solution with true params', color="#e84444", linewidth=1)
plt.plot(t_data, objective_function_analytic.x_analytic(*analytical_solution), '-', label='Fitted Analytic Solution', color="#85f455", linewidth=1)
plt.plot(t_data, objective_function_numerical.x_simulate(*numerical_solution), '-', label='Fitted Numerical Solution', color="#4479e8")
plt.legend(facecolor='#24272f',labelcolor="#e7f3f3",frameon=False)
plt.gcf().set_facecolor('#24272f')  # Figure-Hintergrund
plt.gca().set_facecolor('#24272f')  # Achsen-Hintergrund
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color("#e7f3f3") # x-Achse Farbe
plt.gca().spines['left'].set_color("#e7f3f3") # y-Achse Farbe
plt.tick_params(axis='both', colors="#e7f3f3") # Ticks Farbe
plt.show()
```
