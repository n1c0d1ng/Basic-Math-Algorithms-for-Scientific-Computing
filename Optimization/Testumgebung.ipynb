```python
import numpy as np
num_test_data  = 10
t = np.linspace(0, 1, num_test_data+1)
t
```

```python
def logistische_funktion(t, r, K, x0):
    A = (K - x0) / x0
    return K / (1 + A * np.exp(-r * t))

num_test_data = 100
#Parameter
x0 , K , r = 1, 10, 0.5 

# Zeitgitter anlegen 
t_data = np.linspace(0,10, num_test_data) # 0 bis 1 mit 10 testdaten 
x = logistische_funktion(t_data, r, K, x0)
```

```python
x
```

```python
x_data = x + np.random.normal(0,0.5, num_test_data)
x_data
x_data.shape[0]
```

```python
import matplotlib.pyplot as plt
plt.scatter(t_data, x_data, label='Noisy Data')
plt.plot(t_data, x, color='red', label='True Function')
plt.legend()
plt.show()
```

```python
# Euler Verfahreen fr x
h = t_data[1] - t_data[0]  # Schrittweite
 #Parameter 
x0 = 1.0
r = 0.6
K = 11.0

x_simulated = np.zeros(x_data.shape[0])
x_simulated[0] = x0 

# Euler Verfahren
# x' = r*x*(1-x/K) -> x_{n+1} = x_n + r*x_n*(1-x_n/K)*h
for i in range(1, x_data.shape[0]):
    x_simulated[i] = x_simulated[i-1] + r * x_simulated[i-1] * (1 - x_simulated[i-1]/K) * h

plt.scatter(t_data, x_data, label='Noisy Data')
plt.plot(t_data, x_simulated, color='red', label='True Function')
plt.legend()
plt.show()
```
